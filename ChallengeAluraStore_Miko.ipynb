### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mirko1692/Challenge-Alura-Store-Grupo-9.1/main/tienda_1.csv"
url2 = "https://raw.githubusercontent.com/mirko1692/Challenge-Alura-Store-Grupo-9.1/main/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/mirko1692/Challenge-Alura-Store-Grupo-9.1/main/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/mirko1692/Challenge-Alura-Store-Grupo-9.1/main/tienda_4.csv"

t1 = pd.read_csv(url)
t2 = pd.read_csv(url2)
t3 = pd.read_csv(url3)
t4 = pd.read_csv(url4)

t1.head()
```

#1. Análisis de facturación



```python
ingreso_t1 = t1["Precio"].sum()
ingreso_t2 = t2["Precio"].sum()
ingreso_t3 = t3["Precio"].sum()
ingreso_t4 = t4["Precio"].sum()
```

```python
print(ingreso_t1)
print(ingreso_t2)
print(ingreso_t3)
print(ingreso_t4)
```

# 2. Ventas por categoría

```python
ventas_t1 = t1["Categoría del Producto"].value_counts()
ventas_t2 = t2["Categoría del Producto"].value_counts()
ventas_t3 = t3["Categoría del Producto"].value_counts()
ventas_t4 = t4["Categoría del Producto"].value_counts()
```

```python
print(ventas_t1)
print(ventas_t2)
print(ventas_t3)
print(ventas_t4)
```

# 3. Calificación promedio de la tienda


```python
valoracion_t1 = t1["Calificación"].mean()
valoracion_t2 = t2["Calificación"].mean()
valoracion_t3 = t3["Calificación"].mean()
valoracion_t4 = t4["Calificación"].mean()
```

```python
print(round(valoracion_t1, 3))
print(round(valoracion_t2, 3))
print(round(valoracion_t3, 3))
print(round(valoracion_t4, 3))
```

# 4. Productos más y menos vendidos

```python
mas_vendidos_t1 = t1["Producto"].value_counts().head(5)
menos_vendidos_t1 = t1["Producto"].value_counts().tail(5)

mas_vendidos_t2 = t2["Producto"].value_counts().head(5)
menos_vendidos_t2 = t2["Producto"].value_counts().tail(5)

mas_vendidos_t3 = t3["Producto"].value_counts().head(5)
menos_vendidos_t3 = t3["Producto"].value_counts().tail(5)

mas_vendidos_t4 = t4["Producto"].value_counts().head(5)
menos_vendidos_t4 = t4["Producto"].value_counts().tail(5)
```

```python
print("Tienda 1 - Más vendidos:\n", mas_vendidos_t1)
print("\nTienda 1 - Menos vendidos:\n", menos_vendidos_t1)
print("---------------------------------")
print("\nTienda 2 - Más vendidos:\n", mas_vendidos_t2)
print("\nTienda 2 - Menos vendidos:\n", menos_vendidos_t2)
print("---------------------------------")
print("\nTienda 3 - Más vendidos:\n", mas_vendidos_t3)
print("\nTienda 3 - Menos vendidos:\n", menos_vendidos_t3)
print("---------------------------------")
print("\nTienda 4 - Más vendidos:\n", mas_vendidos_t4)
print("\nTienda 4 - Menos vendidos:\n", menos_vendidos_t4)
```

# 5. Envío promedio por tienda

```python
envio_t1 = t1["Costo de envío"].mean()
envio_t2 = t2["Costo de envío"].mean()
envio_t3 = t3["Costo de envío"].mean()
envio_t4 = t4["Costo de envío"].mean()
```

```python
print(round(envio_t1, 3))
print(round(envio_t2, 3))
print(round(envio_t3, 3))
print(round(envio_t4, 3))
```

## **Visualización**

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title("Ingreso total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso Total")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
print()

# Gráfico de pastel — Distribución de categorías Tienda 1
plt.figure(figsize=(7,8))
ventas_t1.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Distribución de categorías - Tienda 1")
plt.ylabel("")  # quita el texto del eje Y
plt.show()
print()

# Gráfico de pastel — Distribución de categorías Tienda 2
plt.figure(figsize=(7,8))
ventas_t2.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Distribución de categorías - Tienda 2")
plt.ylabel("")  # quita el texto del eje Y
plt.show()
print()

# Gráfico de dispersión — Precio vs Costo de Envío Tienda 1
plt.figure(figsize=(8,5))
plt.scatter(t1["Precio"], t1["Costo de envío"], alpha=0.5, color='red')
plt.title("Precio vs Costo de envío - Tienda 1")
plt.xlabel("Precio")
plt.ylabel("Costo de envío")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Gráfico de dispersión — Precio vs Costo de Envío Tienda 2
plt.figure(figsize=(8,5))
plt.scatter(t2["Precio"], t2["Costo de envío"], alpha=0.5, color='orange')
plt.title("Precio vs Costo de envío - Tienda 2")
plt.xlabel("Precio")
plt.ylabel("Costo de envío")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
